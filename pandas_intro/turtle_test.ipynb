```python
from jupyturtle import make_turtle, left, forward
import math
```

```python
def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)


def arc(radius, angle, n=30):
    arc_length = 2 * math.pi * radius * angle / 360
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)
```

```python
def fibbonaci_spiral(n, radius=1, start=[1, 1]):
    fib = start
    if n > len(start):
        for i in range(len(start), n + 1):
            fib.append(fib[i - 1] + fib[i - 2])
    else:
        fib = start[:n]

    for i in fib:
        arc(i * radius, 90)
    # arc(radius, 180)
    # for i in range(n):
```
